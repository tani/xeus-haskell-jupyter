```haskell
import XHaskell.Display
```

```haskell
-- 使用例のためのデータ定義
newtype HTMLString = HTMLString String

-- HTMLString を Display のインスタンスにする
instance Display HTMLString where
  display (HTMLString html) = DisplayData {
      mimeType = "text/html",
      content  = html
  }

-- 通常の String をプレーンテキストとして表示する場合
instance Display String where
  display str = DisplayData "text/plain" str
```

```haskell
putStr $ show (display (HTMLString "<p style=\"color: red\">foo</p>"))
```

```haskell
show (display (HTMLString "<p style=\"color: red\">foo</p>"))
```
